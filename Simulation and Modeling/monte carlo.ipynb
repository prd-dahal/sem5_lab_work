```python
import random 
import numpy as np
import matplotlib.pyplot as plt
x=[]
y=[]
for i in range(1000):
    x.append(random.uniform(-1,1))
    y.append(random.uniform(-1,1))
plt.scatter(x,y)
x_circle=[]
y_circle=[]
for i in range(1000):
    if((np.square(x[i])+np.square(y[i]))<=1):
        x_circle.append(x[i])
        y_circle.append(y[i])
plt.scatter(x_circle,y_circle)
```
